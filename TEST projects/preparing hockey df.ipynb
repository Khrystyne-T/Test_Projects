```python
test = df.sort_values('game_id')[['game_id',
                                    'away_team_id',
                                    'home_team_id',
                                    'team_id',
                                    'teamName']]
```

```python
#combine team name dups into one col:
com = test.groupby(['game_id', 'away_team_id','home_team_id'])['teamName'].apply(', '.join).reset_index()
```

```python
#test:
com.head()
```

```python
# Split teamName into two seperate cols:
com['HomeTeamName'] = com['teamName'].str.split(',').str.get(0)
com['AwayTeamName'] = com['teamName'].str.split(',').str.get(1)
```

```python
com.head()
```

```python
#rename and drop before merge or else there will be dup cols:
com.teamName.rename('teamNamesList', inplace = True)
com.drop(['away_team_id', 'home_team_id'], axis = 1, inplace = True)
```
